```julia
colnames=Symbol[]
for i in string.(names(df))
    push!(colnames,Symbol(replace(replace(replace(strip(i)," "=> ""),"-"=>""),"/"=> "_")))
end
rename!(df,colnames);
```

```julia
df
```

```julia
using Plots
using StatsPlots
boxplot(df.Lifeexpectancy , title = "Box Plot - Life Expectancy" , ylabel = "Life Expectancy (Years) " , legend = false)
```

```julia
using MLBase
using Lathe
first_percentile=percentile(df.Lifeexpectancy,25)
iqr_value=iqr(df.Lifeexpectancy)
df=df[df.Lifeexpectancy .> (first_percentile - 1.5*-iqr_value),:];
```

```julia

scatter(df.AdultMortality, df.Lifeexpectancy, title="scatter Plot - Life Expectancy", ylabel="Life Expectancy (years)", legand=false)
```

```julia
density(df.Lifeexpectancy, title = "Box Plot - Life Expectancy", ylabel = "Life Expectancy (Years) ", legend = false)
```
